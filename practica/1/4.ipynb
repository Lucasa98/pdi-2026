Para copiar imágenes es necesario utilizar la función `copy()`, ya que la instrucción `=` asigna la misma dirección dem

Si quiere crear una imagen a partir de otra, puede utilizar:

```python
import cv2
import numpy as np

img = cv2.imread("OpenCV-Logo.png")
img_nueva = np.zeros(img.shape, dtype=img.dtype)
# o imagen_nueva = np.zeros_like(imagen)
```

Para cargar una imagen en escala de grises

```python
import matplotlib.pyplot as plt

img_gray = cv2.imread("OpenCV-Logo.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap="gray", vmin=0, vmax=255)
```

Para generar una imagen de grises a partir de una BGR

```python
img_gray2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray2, cmap="gray", vmin=0, vmax=255)
```
