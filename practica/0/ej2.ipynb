# 2 - Hello World

leer imagen, mostrar dimensiones y tipo de dato

```python
import cv2

imagen = cv2.imread("OpenCV-Logo.png")
print("Dimensiones de la imagen: ", imagen.shape)
print("Tipo de dato de la imagen: ", imagen.dtype)
```

mostrar la imagen en una ventana hasta que se presiona una tecla y destruir la ventana

```python
cv2.imshow("Hello World", imagen)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
